# Covid Home Advantage Analysis

## Table showing the difference between points gained at home and away (same with xPoints) from 2014 to 2024 for the 6 championships

### Step 1 : Load and extract the required data

```python
import pandas as pd

data = pd.read_csv('../data/understat-football-matches.csv')

# Filter the data to only include matches from 2014 to 2024 seasons
data = data[(data['season'] >= 2014) & (data['season'] <= 2024)]

# Filter the data to only include the championships, season, h_a, pts and xpts columns
data = data[['championship', 'season', 'h_a', 'pts', 'xpts']]

data.head()
```

### Step 2 : Calculate the difference between points gained at home and away for each championship and season

```python
# Total points gained at home and away for each championship and season
diff_pts = data.groupby(['championship', 'season', 'h_a'])['pts'].sum().unstack().fillna(0)
diff_pts['diff_pts'] = diff_pts['h'] - diff_pts['a']
diff_pts = diff_pts.reset_index()

# Same with xPoints
diff_xpts = data.groupby(['championship', 'season', 'h_a'])['xpts'].sum().unstack().fillna(0)
diff_xpts['diff_xpts'] = diff_xpts['h'] - diff_xpts['a']

# Combine the two dataframes by merging on championship, season with points and xPoints differences
data_diff = pd.merge(diff_pts, diff_xpts, on=['championship', 'season'])
```

### Step 3 : Display the table showing the difference

```python
import pandas as pd

df = data_diff.rename(columns={'championship': 'LEAGUE', 'season': 'YEAR', 'diff_pts': 'DIFF_POINTS_HOMEAWAY', 'diff_xpts': 'DIFF_XPOINTS_HOMEAWAY'})

# Ensure that we're working with only the relevant columns
df = df[['LEAGUE', 'YEAR', 'DIFF_POINTS_HOMEAWAY', 'DIFF_XPOINTS_HOMEAWAY']]

# Round the values to integers
df['DIFF_POINTS_HOMEAWAY'] = df['DIFF_POINTS_HOMEAWAY'].round(0).astype(int)
df['DIFF_XPOINTS_HOMEAWAY'] = df['DIFF_XPOINTS_HOMEAWAY'].round(0).astype(int)

# Use Styler.bar to add horizontal bars to DIFF_POINTS_HOMEAWAY and DIFF_XPOINTS_HOMEAWAY
def add_bars_to_table(df):
    # Apply the bar style to the columns
    styled_df = df.style.bar(subset=['DIFF_POINTS_HOMEAWAY'], color=['red', 'lightgreen'], width=100, align='zero') \
                        .bar(subset=['DIFF_XPOINTS_HOMEAWAY'], color=['red', 'lightgreen'], width=100, align='zero') \
                        .set_table_styles([
                            {'selector': 'thead th', 'props': [('background-color', '#f2f2f2')]},
                            {'selector': 'tbody td', 'props': [('padding', '8px')]}, 
                            {'selector': 'table', 'props': [('border-collapse', 'collapse')]},
                            {'selector': 'td', 'props': [('min-width', '80px')]}
                        ]) \
                        .hide(axis='index')
    
    return styled_df

# Display the styled table
styled_table = add_bars_to_table(df)
styled_table
```

### Conclusion

- With the adding of the seasons after 2020, we still have the same results. 

## Plotting the evolution of mean gained points (per match) for all seasons from 2014 and 2024

### Step 1 : Load and extract the required data

```python
import pandas as pd

data = pd.read_csv('../data/understat-football-matches.csv')

# We only need the points, xPoints, result_away, xPointsAway, championships and season columns
data = data[['pts', 'xpts', 'h_a', 'championship', 'season']]

# Filter the data to only include matches from 2014 to 2024
data = data[(data['season'] >= 2014) & (data['season'] <= 2024)]

data.head()
```

### Step 2 : Calculate the mean gained points per match for each season


```python
# Calculate the mean gained points and expected points per match for each season by grouping by championship, season, and home/away
mean_stats = data.groupby(['championship', 'season', 'h_a']).agg(
    mean_pts=('pts', 'mean'),
    mean_xpts=('xpts', 'mean')
).unstack().fillna(0)

# Rename columns for clarity
mean_stats.columns = ['result_away', 'result', 'xPointsAway', 'xPoints']

# Reset index to make 'championship' and 'season' columns
mean_stats.reset_index(inplace=True)

mean_stats.head()
```

### Step 3 : Plot the evolution of mean gained points per match for all seasons from 2014 to 2020


```python
import matplotlib.pyplot as plt

# Plot the evolution of mean gained points per match for all seasons

for championship in mean_stats['championship'].unique():
    champ_data = mean_stats[mean_stats['championship'] == championship]
 
    plt.figure(figsize=(12, 6))
    plt.plot(champ_data['season'], champ_data['result'], label='result', marker='o', color='blue')
    plt.plot(champ_data['season'], champ_data['xPoints'], label='xPoints', marker='o', color='orange')
    plt.plot(champ_data['season'], champ_data['result_away'], label='result_away', marker="o", color='green')
    plt.plot(champ_data['season'], champ_data['xPointsAway'], label='xPointsAway', marker='o', color='red')
  
    plt.xlabel('Season')
    plt.ylabel('Mean Points per Match')
    plt.title('Evolution of Mean Gained Points per Match (2014-2024) - ' + championship)
    plt.legend()
    plt.grid(True)
    plt.show()
```

### Conclusion

- With the new data, we still have the have the same kind of results for the later seasons comparatively to the 2020 season.

## Paired samples t-test to assess differences between home and away matches, considering either actual result, xPoints, or xG

### Step 1 : Load and extract the required data

```python
from scipy.stats import ttest_rel
import numpy as np
import pandas as pd

data = pd.read_csv('../data/understat-football-matches.csv')

data = data[(data['season'] >= 2014) & (data['season'] <= 2024)]

data = data[['championship', 'season', 'pts', 'xpts', 'xG', 'h_a']]

data.head()
```

### Step 2 : Perform the t-test for each championship and season

```python
ttest_results = []

for (championship, season), group in data.groupby(['championship', 'season']):
    home_data = group[group['h_a'] == 'h']
    away_data = group[group['h_a'] == 'a']
    
    if len(home_data) == len(away_data):

        ttest_points = ttest_rel(home_data['pts'], away_data['pts'])
        
        ttest_xpoints = ttest_rel(home_data['xpts'], away_data['xpts'])
        
        ttest_xg = ttest_rel(home_data['xG'], away_data['xG'])
        
        def compute_cohens_d(a, b):
            diff = a.mean() - b.mean()
            pooled_std = np.sqrt(((len(a) - 1) * a.std()**2 + (len(b) - 1) * b.std()**2) / (len(a) + len(b) - 2))
            return diff / pooled_std if pooled_std != 0 else np.nan
        
        cohens_d_points = compute_cohens_d(home_data['pts'], away_data['pts'])
        cohens_d_xpoints = compute_cohens_d(home_data['xpts'], away_data['xpts'])
        cohens_d_xg = compute_cohens_d(home_data['xG'], away_data['xG'])
        
        # Append results to the list
        ttest_results.append({
            'league': championship,
            'season': season,
            'ttest-result': ttest_points.statistic,
            'ttest-result-pvalue': ttest_points.pvalue,
            'result-cohend': cohens_d_points,
            'ttest-xpoint': ttest_xpoints.statistic,
            'ttest-xpoints-pvalue': ttest_xpoints.pvalue,
            'xpoints-cohend': cohens_d_xpoints,
            'ttest-xg': ttest_xg.statistic,
            'ttest-xg-pvalue': ttest_xg.pvalue,
            'xg-cohend': cohens_d_xg
        })

ttest_df = pd.DataFrame(ttest_results)
```

### Step 3 : Format the results Table


```python
# Highlight p-values < 0.05 and Cohen's d thresholds
def highlight_pval(val):
    return "background-color: red" if val < 0.05 else ""

def highlight_cohens_d(val):
    if val < 0.2:
        return "background-color: yellow"
    elif 0.2 <= val < 0.5:
        return "background-color: orange"
        
        return "background-color: red"
    return ""

styled_ttest_df = ttest_df.style.applymap(highlight_pval, subset=['ttest-result-pvalue', 'ttest-xpoints-pvalue', 'ttest-xg-pvalue']) \
                                       .applymap(highlight_cohens_d, subset=['result-cohend', 'xpoints-cohend', 'xg-cohend'])

styled_ttest_df
```

### Conclusion

- Here, with another statistical test and the addition of new data, we can still draw the same conclusion as the original paper.

## Permutation test to assess differences between seasons (in terms of actual results home and xPoints home)

### Step 1 : Load and extract the required data

```python
import pandas as pd

df = pd.read_csv('../data/understat-football-matches.csv')

df = df[(df['season'] >= 2014) & (df['season'] <= 2024)]

df = df[['championship', 'season', 'xpts', 'h_a']]

# Filter the data to only include home
df = df[df['h_a'] == 'h']
```

### Step 2 : Perform the permutation test for each championship

```python
import numpy as np
import pandas as pd

# Fonction pour effectuer un test de permutation
def permutation_test(data1, data2, n_permutations=1000, statistic=np.mean):
    # Calculer la statistique observée
    observed_stat = np.abs(statistic(data1) - statistic(data2))
    
    # Combiner les deux groupes
    combined = np.concatenate([data1, data2])
    
    # Permutations
    perm_stats = []
    for _ in range(n_permutations):
        np.random.shuffle(combined)
        perm_stats.append(
            np.abs(statistic(combined[:len(data1)]) - statistic(combined[len(data1):]))
        )
    
    # Calculer la p-value
    p_value = np.mean(perm_stats >= observed_stat)
    return observed_stat, p_value

# Fonction pour appliquer le test de permutation à chaque paire de saisons
def compute_permutation_by_league(league, df, n_permutations=1000):
    league_data = df[df['championship'] == league]
    seasons = sorted(league_data['season'].unique())
    
    p_values = []
    # Effectuer le test pour chaque paire de saisons
    for season1 in seasons:
        row = []
        for season2 in seasons:
            if season1 >= season2:
                row.append(np.nan)  # On remplit les parties inutiles avec NaN
                continue
            
            data1 = league_data[league_data['season'] == season1]['xpts']
            data2 = league_data[league_data['season'] == season2]['xpts']
            
            _, p_value = permutation_test(data1, data2, n_permutations=n_permutations)
            row.append(p_value)
        p_values.append(row)
    
    # Créer un DataFrame pour les p-values
    p_value_df = pd.DataFrame(
        p_values,
        columns=[str(s) for s in seasons],
        index=[str(s) for s in seasons]
    )
    
    # Appliquer un style pour mettre en évidence les p-values significatives
    def highlight(p_value):
        if not np.isnan(p_value) and p_value < 0.05:
            return 'color: red'
        return ''
    
    styled_df = p_value_df.style.applymap(highlight)
    return styled_df

# Appliquer le test de permutation pour chaque championnat et afficher le tableau stylisé
leagues = df['championship'].unique()

for league in leagues:
    print(f"\n{league} xPoints Change Significance with Permutation Test")
    styled_table = compute_permutation_by_league(league, df, n_permutations=1000)
    display(styled_table)
```

### Conclusion

The **t-test results** indicate significant differences in football performance metrics (points, xpoints, xG) across most leagues and seasons, with high effect sizes in many cases, especially in earlier years. However, some seasons, like 2019-2020, showed non-significant results, likely due to factors like the COVID-19 pandemic.

Additionally, the **permutation test** complements the t-test by validating the observed differences in xPoints across multiple seasons, offering a robust, distribution-free alternative to confirm or dispute the statistical significance of changes. The alignment between the permutation test and t-test in the results adds further credibility to the conclusions drawn from both analyses.

Overall, both tests highlight meaningful trends in team performance across seasons, with some fluctuations influenced by external factors.

- With the new statistical test, we still get the same conclusion and the new data confirm this result.
